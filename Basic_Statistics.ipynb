3.  Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?

  - Dispersion refers to the spread or variability of a dataset. It measures how much the values deviate from the central tendency (mean, median, or mode). A dataset with low dispersion has values that are closely clustered, while high dispersion indicates that the values are widely spread out.

  Why is Dispersion Important?
Helps understand the reliability of the mean.

Identifies whether data points are consistent or highly variable.

Crucial for making comparisons between different datasets.

Measures of Dispersion
Two key measures of dispersion are variance and standard deviation, both of which quantify how much each data point deviates from the mean.

5.  Discuss the role of random sampling in making inferences about populations.
    - Random sampling is a method of selecting a subset of individuals from a population where each member has an equal chance of being chosen. This technique is essential in statistics because it ensures that the sample fairly represents the entire population, allowing for accurate inferences.

6.  Explain the concept of skewness and its types. How does skewness affect the interpretation of data?

 - Skewness is a measure of the asymmetry of a datasetâ€™s distribution. It indicates whether the data values are symmetrically distributed around the mean or if they have a longer tail on one side.

A perfectly symmetrical dataset has zero skewness, while an asymmetrical dataset is either positively or negatively skewed.

1. Symmetric Distribution (Zero Skewness)
The left and right sides of the distribution are mirror images.

Mean, median, and mode are equal.

Example: Heights of people in a population.

2. Positive Skewness (Right-Skewed)
The right tail is longer, meaning there are more high-value outliers.

Mean > Median > Mode.

Example: Income distribution (a few very high salaries pull the mean upwards).

3. Negative Skewness (Left-Skewed)
The left tail is longer, meaning there are more low-value outliers.

Mean < Median < Mode.

Example: Age of retirement (most people retire around the same age, but a few retire very early).

How Skewness Affects Data Interpretation
Impact on Measures of Central Tendency:

In symmetrical distributions, the mean, median, and mode are nearly the same.

In skewed distributions, the mean is pulled toward the longer tail, making the median a better representation of the center.

Effect on Decision-Making:

In right-skewed data (e.g., salaries), the mean overestimates the typical value.

In left-skewed data (e.g., test scores with many low outliers), the mean underestimates the central tendency.

Influence on Statistical Analysis:

Many statistical tests (like t-tests and ANOVA) assume normality (zero skewness). If data is skewed, transformation techniques (e.g., log transformation) may be needed.

Skewness affects risk analysis in financeâ€”right-skewed distributions indicate potential for extreme profits, while left-skewed distributions indicate risk of extreme losses.

11.  Explain what a random variable is and differentiate between discrete and continuous random variables.
   - A random variable is a numerical value that represents the outcome of a random experiment. It assigns a real number to each possible outcome in a probability experiment.

Types of Random Variables
1. Discrete Random Variable
A discrete random variable takes on a finite or countable number of distinct values. It is typically associated with counting (e.g., number of people, number of defects).

Examples:

Number of students in a class.

Number of heads in 10 coin flips.

Number of emails received in an hour.

These values are countable and not continuous.

2. Continuous Random Variable
A continuous random variable can take on any value within a given range, typically representing measurements. The number of possible values is uncountable (infinite within an interval).

Examples:

Height of a person (e.g., 170.5 cm, 171.2 cm).

Time taken to finish a race (e.g., 9.85 seconds).

Temperature in a city (e.g., 23.4Â°C, 23.45Â°C).

These values can take on any real number within an interval and are not just whole numbers.
